```python
import pandas as pd
```

```python
deaths_states = pd.read_csv("../epidemic/deaths_state_21.csv")
death_states_monthly = deaths_states.copy()
```

```python
deaths_states.head(1000)
```

```python
#Converts the data to deaths_new monthly average to avoid 0 values, same for cumul_full
deaths_states_monthly = deaths_states.groupby(['date','state'])[['deaths_new', 'cumul_full','population','cumul_partial']].mean().reset_index()
deaths_states_monthly['date'] = pd.to_datetime(deaths_states_monthly['date'], format = "%d/%m/%Y")
deaths_states_monthly['date'] = deaths_states_monthly['date'].dt.to_period('M')
deaths_states_monthly.head(30)
```

```python
#Finding out percentage cummulation for each state
deaths_states_monthly["cumul_full"] = pd.to_numeric(deaths_states_monthly['cumul_full'])
deaths_states_monthly["population"] = pd.to_numeric(deaths_states_monthly['population'])
deaths_states_monthly["percentage_vax_partial"] = (deaths_states_monthly['cumul_partial']/deaths_states_monthly['population']) * 100
deaths_states_monthly["percentage_vax_full"] = (deaths_states_monthly['cumul_full']/deaths_states_monthly['population']) * 100
deaths_states_monthly
```

```python
deaths_states_monthly.to_csv("deaths_vax_states_monthly.csv")
```
